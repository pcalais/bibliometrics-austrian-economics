```python
import math

def map_year_to_book_hayek(year):
    if math.isnan(year):
        return 'unknown-year'
    
    year = int(year)
    if year == 1945:
        return "Use of Knowledge in Society"
    if year == 1948:
        return "Individualism and Economic Order"
    if year == 1952:
        return "The Counter-Evolution of Science"
    if year == 1960:
        return 'The Constitution of Liberty'
    if year == 1978:
        return 'New Studies in Philosophy, Politics, Economics and the History of Ideas'
    if year == 1973:
        return 'Law, Legislation and Liberty – Volume 1: Rules and Order'
    if year == 1967:
        return 'Studies in Philosophy, Politics and Economics'
    if year == 1937:
        return 'Economics and Knowledge'
    
    return "unknown"



# Aplicando ao DataFrame
hayek_citations_df['book'] = hayek_citations_df['year'].apply(map_year_to_book_hayek)
```

```python
import pandas as pd

# Lê o arquivo assumindo que é separado por vírgulas
human_action_page_ranges_df = pd.read_csv("chapter-ranges/human-action-page-to-chapter.txt")

human_action_page_ranges_df.head()
```

```python
def get_human_action_chapter_by_page(page):
    row = human_action_page_ranges_df[
        (human_action_page_ranges_df['start_page'] <= page) &
        (human_action_page_ranges_df['end_page'] >= page)
    ]
    if not row.empty:
        return int(row.iloc[0]['chapter'])
    else:
        return "WHOLE"  # ou você pode retornar "Desconhecido", -1, etc.

assert get_human_action_chapter_by_page(111) == 6
assert get_human_action_chapter_by_page(None) == 'WHOLE'
```

```python
mises_citations_df['human_action_chapter_by_ref_page'] = mises_citations_df['page'].apply(get_human_action_chapter_by_page)
mises_citations_df['human_action_chapter_number'] = mises_citations_df['human_action_chapter_by_ref_page']

```

```python
chapter_to_part_df = pd.read_csv("chapter-ranges/chapter-to-part.csv")

chapter_to_part_df.head()
```

```python
import pandas as pd

# 🔧 Força os campos de join para string em ambos os DataFrames
mises_citations_df["human_action_chapter_number"] = (
    mises_citations_df["human_action_chapter_number"].astype(str)
)

chapter_to_part_df["human_action_chapter_number"] = (
    chapter_to_part_df["human_action_chapter_number"].astype(str)
)

# 🔗 Join usando merge
mises_citations_df = pd.merge(
    mises_citations_df,
    chapter_to_part_df,
    on="human_action_chapter_number",
    how="left"
)

# ✅ Existem pelo menos 2 valores de part distintos?
assert (
    mises_citations_df["human_action_part_number"]
    .nunique(dropna=True) >= 2
), "human_action_part_number não possui pelo menos dois valores distintos"

```

```python
# ===== 1. Salvar por sentença =====
mises_citations_df.to_csv("../data/processed/mises_sentence_citations.csv", index=False)
hayek_citations_df.to_csv("../data/processed/hayek_sentence_citations.csv", index=False)


# ===== Função auxiliar: lista única preservando ordem =====
def unique_list(x):
    seen = set()
    result = []
    for item in x:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result


# ===== 2. Função para agregação por parágrafo =====
def aggregate_paragraph(df):
    agg_dict = {
        # mantém o parágrafo (único por paragraph_id)
        "paragraph": "first",
        "paragraph_keywords": "first",
        "sentence_keywords": "first",
        
        # mantém sentence_ids únicos
        "sentence_id": unique_list
    }
    
    # adiciona dinamicamente se a coluna existir
    if "human_action_chapter_name" in df.columns:
        agg_dict["human_action_chapter_name"] = unique_list
    if "human_action_chapter_number" in df.columns:
        agg_dict["human_action_chapter_number"] = unique_list
    if "human_action_part_name" in df.columns:
        agg_dict["human_action_part_name"] = unique_list
    if "human_action_part_number" in df.columns:
        agg_dict["human_action_part_number"] = unique_list

    return (
        df
        .sort_values("sentence_seq_number")  # garante ordem correta
        .groupby("paragraph_id")
        .agg(agg_dict)
        .reset_index()
    )


# ===== 3. Aplicar agregação =====
mises_paragraph_df = aggregate_paragraph(mises_citations_df)
mises_paragraph_df.to_csv("../data/processed/mises_paragraph_citations.csv", index=False)


# ===== 4. Salvar por parágrafo =====
hayek_paragraph_df = aggregate_paragraph(hayek_citations_df)
hayek_paragraph_df.to_csv("../data/processed/hayek_paragraph_citations.csv", index=False)
```
