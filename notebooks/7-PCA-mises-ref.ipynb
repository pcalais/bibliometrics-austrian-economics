```python
import pandas as pd

# Remove linhas sem chapter
df = mises_refs_df.dropna(subset=["paper_id", "human_action_part_name"])


# Cria matriz paper × chapter (binária)
pca_input_df = (
    pd.crosstab(
        index=df["paper_id"],
        columns=df["human_action_part_name"]
    )
    .astype(int)
)

pca_input_df.head()
```

```python
import pandas as pd

# Remove linhas sem paper ou parte
df = mises_refs_df.dropna(subset=["paper_id", "human_action_part_name"])

# Matriz paper × part (contagem absoluta)
count_df = pd.crosstab(
    index=df["paper_id"],
    columns=df["human_action_part_name"]
)

# Normaliza por linha para obter percentuais (soma = 100%)
pca_input_df = count_df.div(count_df.sum(axis=1), axis=0) * 100

pca_input_df.head()
```

```python
# PADRONIZAR

from sklearn.preprocessing import StandardScaler

X = pca_input_df.values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# RODAR

from sklearn.decomposition import PCA

# Escolha inicial: 2 componentes (exploratória)
pca = PCA(n_components=7, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
# SCORES

pca_scores_df = pd.DataFrame(
    X_pca,
    index=pca_input_df.index,
    columns=["PC1", "PC2", "PC3", "PC4", "PC5", "PC6", "PC7"]
)

pca_scores_df.head()
```

```python
# VARIANCIA

explained_variance_df = pd.DataFrame({
    "PC": ["PC1", "PC2", "PC3", "PC4", "PC5", "PC6", "PC7"],
    "explained_variance_ratio": pca.explained_variance_ratio_,
    "explained_variance_cumulative": pca.explained_variance_ratio_.cumsum()
})

explained_variance_df
```

```python
# LOADINGS

loadings_df = pd.DataFrame(
    pca.components_.T,
    index=pca_input_df.columns,
    columns=["PC1", "PC2", "PC3", "PC4", "PC5", "PC6", "PC7"]
)

loadings_df.sort_values("PC1", ascending=False).head(100)
```
