```python
import pandas as pd

# LÃª o arquivo assumindo que Ã© separado por vÃ­rgulas
human_action_page_ranges_df = pd.read_csv("chapter-ranges/human-action-page-to-chapter.txt")

human_action_page_ranges_df.head()
```

```python
def get_human_action_chapter_by_page(page):
    row = human_action_page_ranges_df[
        (human_action_page_ranges_df['start_page'] <= page) &
        (human_action_page_ranges_df['end_page'] >= page)
    ]
    if not row.empty:
        return int(row.iloc[0]['chapter'])
    else:
        return "WHOLE"  # ou vocÃª pode retornar "Desconhecido", -1, etc.

assert get_human_action_chapter_by_page(111) == 6
assert get_human_action_chapter_by_page(None) == 'WHOLE'
```

```python
mises_refs_df['human_action_chapter_by_ref_page'] = mises_refs_df['page'].apply(get_human_action_chapter_by_page)
mises_refs_df['human_action_chapter_number'] = mises_refs_df['human_action_chapter_by_ref_page']


```

```python
chapter_to_part_df = pd.read_csv("chapter-ranges/chapter-to-part.csv")

chapter_to_part_df.head()
```

```python
import pandas as pd

# ðŸ”§ ForÃ§a os campos de join para string em ambos os DataFrames
mises_refs_df["human_action_chapter_number"] = (
    mises_refs_df["human_action_chapter_number"].astype(str)
)

chapter_to_part_df["human_action_chapter_number"] = (
    chapter_to_part_df["human_action_chapter_number"].astype(str)
)

# ðŸ”— Join usando merge
mises_refs_df = pd.merge(
    mises_refs_df,
    chapter_to_part_df,
    on="human_action_chapter_number",
    how="left"
)

# âœ… Existem pelo menos 2 valores de part distintos?
assert (
    mises_refs_df["human_action_part_number"]
    .nunique(dropna=True) >= 2
), "human_action_part_number nÃ£o possui pelo menos dois valores distintos"
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter

# 1ï¸âƒ£ Sentence_ids relevantes (vindos do mises_refs_df)
sentence_ids = set(mises_refs_df["sentence_id"].dropna())

# 2ï¸âƒ£ Filtra refs_df apenas para essas sentenÃ§as
filtered_refs = refs_df[
    refs_df["sentence_id"].isin(sentence_ids)
].copy()

# Remove autores nulos
filtered_refs = filtered_refs.dropna(subset=["author"])

# 3ï¸âƒ£ Agrupa autores por sentence_id
authors_by_sentence = (
    filtered_refs
    .groupby("sentence_id")["author"]
    .apply(lambda x: sorted(set(x)))
)

# 4ï¸âƒ£ Gera pares de autores por sentenÃ§a
edge_counter = Counter()

for authors in authors_by_sentence:
    if len(authors) < 2:
        continue
    for a, b in combinations(authors, 2):
        edge_counter[(a, b)] += 1

# 5ï¸âƒ£ Converte para DataFrame
edges_df = pd.DataFrame(
    [(a, b, w) for (a, b), w in edge_counter.items()],
    columns=["source", "target", "weight"]
)

# ðŸ”½ ORDENA por peso (decrescente)
edges_df = edges_df.sort_values(
    by="weight",
    ascending=False
)

# (opcional) remove arestas muito fracas
edges_df = edges_df[edges_df["weight"] >= 1]

# 6ï¸âƒ£ Exporta no formato do VOSviewer
edges_df.to_csv(
    "../data/processed/network/mises_network.txt",
    sep="\t",
    index=False,
    header=False,
    encoding="utf-8"
)

print(f"Arquivo mises_network.txt gerado com {len(edges_df)} arestas, ordenado por weight.")
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter


def gera_rede_por_page(
    refs_df,
    sentence_ids,
    page_is_nan=True,
    min_weight=1
):
    """
    Gera uma rede de co-citaÃ§Ã£o por sentence_id.

    page_is_nan=True  -> usa apenas refs com page NaN
    page_is_nan=False -> usa apenas refs com page != NaN
    """

    # 1ï¸âƒ£ Filtra apenas sentence_ids relevantes
    filtered_refs = refs_df[
        refs_df["sentence_id"].isin(sentence_ids)
    ].copy()

    # 2ï¸âƒ£ Filtra por presenÃ§a ou ausÃªncia de page
    if page_is_nan:
        filtered_refs = filtered_refs[filtered_refs["page"].isna()]
    else:
        filtered_refs = filtered_refs[filtered_refs["page"].notna()]

    # 3ï¸âƒ£ Remove autores nulos
    filtered_refs = filtered_refs.dropna(subset=["author"])

    # 4ï¸âƒ£ Agrupa autores Ãºnicos por sentence_id
    authors_by_sentence = (
        filtered_refs
        .groupby("sentence_id")["author"]
        .apply(lambda x: sorted(set(x)))
    )

    # 5ï¸âƒ£ Gera pares de autores por sentenÃ§a
    edge_counter = Counter()

    for authors in authors_by_sentence:
        if len(authors) < 2:
            continue
        for a, b in combinations(authors, 2):
            edge_counter[(a, b)] += 1

    # 6ï¸âƒ£ Converte para DataFrame
    edges_df = pd.DataFrame(
        [(a, b, w) for (a, b), w in edge_counter.items()],
        columns=["source", "target", "weight"]
    )

    # 7ï¸âƒ£ Ordena por peso (decrescente)
    edges_df = edges_df.sort_values(
        by="weight",
        ascending=False
    )

    # 8ï¸âƒ£ Remove arestas fracas (opcional)
    edges_df = edges_df[edges_df["weight"] >= min_weight]

    return edges_df


# =========================================================
# ðŸš€ EXECUÃ‡ÃƒO
# =========================================================

# Sentence_ids relevantes (vindos do mises_refs_df)
sentence_ids = set(mises_refs_df["sentence_id"].dropna())

# ðŸ”¹ REDE 1 â€” citaÃ§Ãµes SEM pÃ¡gina
edges_no_page = gera_rede_por_page(
    refs_df=refs_df,
    sentence_ids=sentence_ids,
    page_is_nan=True
)

edges_no_page.to_csv(
    "../data/processed/network/mises_network_no_page.txt",
    sep="\t",
    index=False,
    header=False,
    encoding="utf-8"
)

print(
    f"âœ” Rede SEM pÃ¡gina gerada com {len(edges_no_page)} arestas."
)


# ðŸ”¹ REDE 2 â€” citaÃ§Ãµes COM pÃ¡gina
edges_with_page = gera_rede_por_page(
    refs_df=refs_df,
    sentence_ids=sentence_ids,
    page_is_nan=False
)

edges_with_page.to_csv(
    "../data/processed/network/mises_network_with_page.txt",
    sep="\t",
    index=False,
    header=False,
    encoding="utf-8"
)

print(
    f"âœ” Rede COM pÃ¡gina gerada com {len(edges_with_page)} arestas."
)
```
