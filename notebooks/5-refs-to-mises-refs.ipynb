```python
def get_human_action_chapter_by_page(page):
    row = human_action_page_ranges_df[
        (human_action_page_ranges_df['start_page'] <= page) &
        (human_action_page_ranges_df['end_page'] >= page)
    ]
    if not row.empty:
        return int(row.iloc[0]['chapter'])
    else:
        return "WHOLE"  # ou vocÃª pode retornar "Desconhecido", -1, etc.

assert get_human_action_chapter_by_page(111) == 6
assert get_human_action_chapter_by_page(None) == 'WHOLE'
```

```python
import pandas as pd
from itertools import combinations

rows = []

# MantÃ©m apenas linhas vÃ¡lidas
df = mises_refs_df.dropna(
    subset=[
        "paper_id",
        "raw",
        "human_action_part_number",
        "sentence_seq_number"
    ]
)

# Agrupa por paper_id
for paper_id, group in df.groupby("paper_id"):
    # Se houver menos de 2 citaÃ§Ãµes, nÃ£o hÃ¡ pares
    if len(group) < 2:
        continue

    # Gera todos os pares possÃ­veis dentro do paper
    for (_, r1), (_, r2) in combinations(group.iterrows(), 2):
        rows.append({
            "paper_id": paper_id,
            "sentence_id_1": r1["sentence_id"],
            "sentence_id_2": r2["sentence_id"],
            "raw_1": r1["raw"],
            "raw_2": r2["raw"],
            "human_action_part_number_1": r1["human_action_part_number"],
            "human_action_part_number_2": r2["human_action_part_number"],
            "distance": abs(
                r1["sentence_seq_number"] - r2["sentence_seq_number"]
            )
        })

# DataFrame final
mises_ref_pair_df = pd.DataFrame(rows)

mises_ref_pair_df.to_csv("../data/processed/mises_ref_pairs.csv")

mises_ref_pair_df
```

```python
import pandas as pd

# Seleciona apenas o que interessa do mises_refs_df
mises_parts = (
    mises_refs_df[
        ["sentence_id", "author", "human_action_part_number"]
    ]
    .dropna(subset=["sentence_id", "human_action_part_number"])
    .copy()
)

# Faz LEFT JOIN em refs_df
refs_df = refs_df.merge(
    mises_parts,
    on=["sentence_id", "author"],
    how="left"
)


def build_author_node(row):
    if (
        row["author"] == "Mises"
        and pd.notna(row["human_action_part_number"])
    ):
        return f"Mises_{(row['human_action_part_number'])}"
    return row["author"]


refs_df = refs_df.copy()
refs_df["author"] = refs_df.apply(build_author_node, axis=1)
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter


def build_cocitation_networks_scope_intensity(
    refs_df: pd.DataFrame,
    unit_col: str,
    min_weight: int = 2
):
    """
    Retorna dois DataFrames:
    1) scope_df     â†’ co-citaÃ§Ã£o sem repetiÃ§Ã£o (presenÃ§a por unidade)
    2) intensity_df â†’ co-citaÃ§Ã£o com repetiÃ§Ã£o (Ãªnfase retÃ³rica)

    unit_col âˆˆ {"sentence_id", "paragraph_id", "paper_id"}
    """

    # Filtra dados vÃ¡lidos
    df = (
        refs_df
        .dropna(subset=[unit_col, "author"])
        .copy()
    )

    # Agrupa autores por unidade (lista completa, com repetiÃ§Ã£o)
    authors_by_unit = (
        df
        .groupby(unit_col)["author"]
        .apply(list)
    )

    # Contadores
    scope_counter = Counter()
    intensity_counter = Counter()

    for authors in authors_by_unit:
        if len(authors) < 2:
            continue

        # ---------- SCOPE (sem repetiÃ§Ã£o) ----------
        unique_authors = sorted(set(authors))
        for a, b in combinations(unique_authors, 2):
            scope_counter[(a, b)] += 1

        # ---------- INTENSITY (com repetiÃ§Ã£o) ----------
        for a, b in combinations(sorted(authors), 2):
            if a != b:
                intensity_counter[(a, b)] += 1

    # Converte para DataFrame
    scope_df = pd.DataFrame(
        [(a, b, w) for (a, b), w in scope_counter.items()],
        columns=["source", "target", "weight"]
    ).sort_values("weight", ascending=False)

    intensity_df = pd.DataFrame(
        [(a, b, w) for (a, b), w in intensity_counter.items()],
        columns=["source", "target", "weight"]
    ).sort_values("weight", ascending=False)

    # Remove arestas fracas
    scope_df = scope_df[scope_df["weight"] >= min_weight]
    intensity_df = intensity_df[intensity_df["weight"] >= min_weight]

    return scope_df, intensity_df
```

```python
# Sentence-level
scope_sent, intensity_sent = build_cocitation_networks_scope_intensity(
    refs_df,
    unit_col="sentence_id"
)

# Paragraph-level
scope_par, intensity_par = build_cocitation_networks_scope_intensity(
    refs_df,
    unit_col="paragraph_id"
)

# Paper-level (âš ï¸ intensidade aqui Ã© conceitualmente irrelevante)
scope_paper, _ = build_cocitation_networks_scope_intensity(
    refs_df,
    unit_col="paper_id"
)
```

```python
import pandas as pd
from collections import defaultdict

UNIT_COL = "sentence_id"   # troque para paragraph_id ou paper_id se quiser

df = (
    refs_df
    .dropna(subset=[UNIT_COL, "author"])
    .copy()
)

# Agrupa autores por unidade discursiva
authors_by_unit = (
    df
    .groupby(UNIT_COL)["author"]
    .apply(set)
)
```

```python
author_mises_counts = defaultdict(lambda: defaultdict(int))

for authors in authors_by_unit:
    mises_nodes = {a for a in authors if a.startswith("Mises_")}
    non_mises = {a for a in authors if not a.startswith("Mises_")}

    if not mises_nodes or not non_mises:
        continue

    for author in non_mises:
        for mises in mises_nodes:
            author_mises_counts[author][mises] += 1
```

```python
author_mises_rows = []

for author, mises_dict in author_mises_counts.items():
    total = sum(mises_dict.values())
    for mises, count in mises_dict.items():
        author_mises_rows.append({
            "author": author,
            "mises_part": mises,
            "count": count,
            "percent": 100 * count / total
        })

author_mises_stats_df = pd.DataFrame(author_mises_rows)

author_mises_stats_df.head()
```

```python
author_mises_percent_table = (
    author_mises_stats_df
    .pivot_table(
        index="author",
        columns="mises_part",
        values="percent",
        fill_value=0
    )
)

# 1ï¸âƒ£ Calcula o total de co-citaÃ§Ãµes por autor
author_total_counts = (
    author_mises_stats_df
    .groupby("author")["count"]
    .sum()
)

# 2ï¸âƒ£ Adiciona como coluna Ã  tabela de percentuais
author_mises_percent_table = author_mises_percent_table.copy()
author_mises_percent_table["count"] = author_mises_percent_table.index.map(
    author_total_counts
)

author_mises_percent_table.head()
```

```python
def rank_authors_by_mises_cocitation(
    refs_df: pd.DataFrame,
    unit_col: str
) -> pd.Series:
    """
    Retorna um Series:
    index = author
    value = ranking (1 = maior co-citaÃ§Ã£o com Mises)
    """

    df = (
        refs_df
        .dropna(subset=[unit_col, "author"])
        .copy()
    )

    # Agrupa autores por unidade
    authors_by_unit = (
        df
        .groupby(unit_col)["author"]
        .apply(set)
    )

    # Conta co-citaÃ§Ãµes Autor Ã— Mises (scope)
    from collections import Counter
    author_counter = Counter()

    for authors in authors_by_unit:
        mises_nodes = {a for a in authors if a.startswith("Mises_")}
        non_mises = {a for a in authors if not a.startswith("Mises_")}

        if not mises_nodes:
            continue

        for author in non_mises:
            author_counter[author] += len(mises_nodes)

    # Converte para ranking
    ranking_df = (
        pd.DataFrame(
            author_counter.items(),
            columns=["author", "count"]
        )
        .sort_values("count", ascending=False)
    )

    ranking_df["rank"] = ranking_df["count"].rank(
        method="min",
        ascending=False
    ).astype(int)

    return ranking_df.set_index("author")["rank"]
```

```python
rank_sentence = rank_authors_by_mises_cocitation(
    refs_df, unit_col="sentence_id"
)

rank_paragraph = rank_authors_by_mises_cocitation(
    refs_df, unit_col="paragraph_id"
)

rank_head = rank_authors_by_mises_cocitation(
    refs_df, unit_col="head_id"
)

rank_paper = rank_authors_by_mises_cocitation(
    refs_df, unit_col="paper_id"
)
```

```python
author_ranking_df = pd.DataFrame({
    "sentence": rank_sentence,
    "paragraph": rank_paragraph,
    "head": rank_head,
    "paper": rank_paper
})

author_ranking_df = author_ranking_df.sort_values(
    #by=["sentence", "paragraph", "paper"],
    by=["paper", "paragraph", "head", "paper"]
)

author_ranking_df.head(200)
```
