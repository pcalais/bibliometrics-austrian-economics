```python
import requests
from bs4 import BeautifulSoup

def download_qjae(url):

    # Fetch the webpage
    response = requests.get(url)

    #print('response', response.text)

    if response.status_code != 200:
        print("Failed to fetch the webpage")
        exit()

    # Parse the page content
    soup = BeautifulSoup(response.text, 'html.parser')

    if ".pdf" in url:
        pdf_response = requests.get(url)
        if pdf_response.status_code == 200:
            with open("downloaded_file.pdf", "wb") as file:
                file.write(pdf_response.content)
                print("PDF downloaded successfully: downloaded_file.pdf")
                return pdf_response.content
        else:
            print("Failed to download the PDF")
            return

    if "link.springer.com" in url:
        print('SPRINGER')
        #links = soup.find_all('a', attrs={'data-track-action': 'download pdf'})
        links = soup.find_all('a', attrs={'data-track-action': lambda x: x in ['download pdf', 'Book download - pdf']})

        if len(links) > 0:
            first_link_href = links[0].get('href')
            #soup = BeautifulSoup(link, 'html.parser')
            #href_value = soup.find('a')['href']
            #print('href_value', first_link_href)

            response = requests.get("https://link.springer.com/" + first_link_href)
        else:
             print(f"[SPRINGER] Failed to download {url}. Link not found.")
             return None

         # Check if the request was successful
        if response.status_code == 200:
            print(f"[SPRINGER] {first_link_href} File successfully downloaded")
            return response.content
        else:
            print(f"[SPRINGER] Failed to download {first_link_href}. Status code: {response.status_code}")
            return None


    # Find the meta tag with the PDF link
    meta_tag = soup.find('meta', attrs={'name': 'citation_pdf_url'})
    if not meta_tag or not meta_tag.get('content'):
        print("PDF link not found")
        exit()

    pdf_url = meta_tag['content']

    # Download the PDF
    pdf_response = requests.get(pdf_url)
    if pdf_response.status_code == 200:
        with open("downloaded_file.pdf", "wb") as file:
            file.write(pdf_response.content)
        print("PDF downloaded successfully: downloaded_file.pdf")
    else:
        print("Failed to download the PDF")
```

```python
import requests

def get_final_url(url):
    try:
        response = requests.head(url, allow_redirects=True)
        return response.url
    except requests.RequestException as e:
        return f"Error: {e}"
```

```python
def is_downloadable(row):
    return "Springer" in row['Publisher'] and "Open Access" in row['Open Access'] or \
    "Review of Austrian Economics" in row['Source title']
```

```python
import pandas as pd
import os
import time


# File paths
csv_file = '../data/raw/scopus.csv'
output_dir = '../data/raw/pdf'

# Ensure the output directory exists
#os.makedirs(output_dir, exist_ok=True)

# Read the dataset
df_to_download = pd.read_csv(csv_file)
df_to_download = df_to_download.fillna('')


df_to_download['is_downloadable'] = df_to_download.apply(is_downloadable, axis=1)
df_to_download = df_to_download.sort_values(by='is_downloadable', ascending=False)


df_download = pd.DataFrame()

df_to_download = df_to_download.sample(frac=1).reset_index(drop=True)


# Iterate over rows and run the command
paper_id = 0
downloaded_papers = 0
for _, row in df_to_download.iterrows():
    doi = row['DOI']
    if pd.notna(doi):  # Ensure DOI is not NaN

        #print(doi)
        clean_doi = doi.replace("/", ".")

        output_file_path = f"{output_dir}/{clean_doi}.pdf" 

        if os.path.exists(output_file_path):
            print(f"{output_file_path} already downloaded")
            continue

        final_url = get_final_url(f"https://doi.org/{doi}")
        print(final_url, paper_id, row['Open Access'])


        if "link.springer.com" in final_url and "Open Access" in row['Open Access']:
            pdf_content = download_qjae(final_url)
            if pdf_content is not None:
                with open(output_file_path, 'wb') as file:
                    file.write(pdf_content)
                    row['download_source'] = 'springer_link'

        else:
            a = 1
            continue
            command = f"scidownl download --doi https://doi.org/{doi} --out {output_file_path} --scihub-url http://sci-hub.se"
            print(command)
            os.system(command)

        row['clean_doi'] = clean_doi
        row['paper_id'] = paper_id
        row['final_url'] = final_url
        downloaded = os.path.exists(output_file_path) and os.path.getsize(output_file_path) > 4096  # 1KB = 1024 bytes
        row['downloaded'] = downloaded
        df_download = pd.concat([df_download, pd.DataFrame([row])], ignore_index=True)

        if downloaded:
            downloaded_papers += 1
            print("DOWNLOADED! # of papers downloaded:", downloaded_papers + "\n")
        else:
            print("ERROR trying to download.\n")    

        paper_id += 1
        time.sleep(1)  # Pause for 1 second
```
