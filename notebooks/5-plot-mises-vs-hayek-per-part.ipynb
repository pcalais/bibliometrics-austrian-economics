```python
import ast
import matplotlib.pyplot as plt
import pandas as pd

# --------------------------------------------------
# 1️⃣ Converter strings para listas reais
# --------------------------------------------------
cols_to_explode = [
    'human_action_chapter_name',
    'human_action_part_name',
    'human_action_part_number',
    'human_action_chapter_number'
]

df = mises_citations_df.copy()

def safe_literal_eval(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return [x]
    return x

for col in cols_to_explode:
    df[col] = df[col].apply(safe_literal_eval)

# --------------------------------------------------
# 2️⃣ Garantir que tudo é lista
# --------------------------------------------------
for col in cols_to_explode:
    df[col] = df[col].apply(
        lambda x: x if isinstance(x, list) else [x]
    )

# --------------------------------------------------
# 3️⃣ Agora SIM explode funciona
# --------------------------------------------------
df = df.explode(cols_to_explode)

# --------------------------------------------------
# 4️⃣ Filtragem
# --------------------------------------------------
filtered_df = df[
    (df['human_action_chapter_name'] != 'not found') &
    (df['human_action_part_number'] != 'WHOLE')
].copy()

# --------------------------------------------------
# 5️⃣ Conversão para inteiro
# --------------------------------------------------
filtered_df['human_action_part_number'] = filtered_df['human_action_part_number'].astype(int)
filtered_df['human_action_chapter_number'] = filtered_df['human_action_chapter_number'].astype(int)

# --------------------------------------------------
# 6️⃣ Cores
# --------------------------------------------------
part_colors = {
    1: 'lightblue',
    2: 'lightgreen',
    3: 'lightcoral',
    4: 'lightgoldenrodyellow',
    5: 'lightsalmon',
    6: 'lightpink',
    7: 'lightgray'
}

filtered_df['bar_color'] = filtered_df['human_action_part_number'].map(part_colors)

# --------------------------------------------------
# 7️⃣ Contagem
# --------------------------------------------------
chapter_counts_sorted = (
    filtered_df
        .groupby(
            [
                'human_action_chapter_name',
                'human_action_chapter_number',
                'human_action_part_number',
                'bar_color'
            ]
        )
        .size()
        .reset_index(name='counts')
        .sort_values(by='human_action_chapter_number', ascending=False)
)

# --------------------------------------------------
# 8️⃣ Plot
# --------------------------------------------------
plt.figure(figsize=(12, 8))
plt.barh(
    chapter_counts_sorted['human_action_chapter_name'],
    chapter_counts_sorted['counts'],
    color=chapter_counts_sorted['bar_color'],
    edgecolor='black'
)

plt.title('Histograma do campo "human_action_chapter" (excluindo WHOLE)')
plt.xlabel('Frequência')
plt.ylabel('Chapter')

plt.tight_layout()
plt.savefig('../reports/human_action_chapters.png')
plt.show()
```

```python
import ast
import matplotlib.pyplot as plt
import pandas as pd

unit = 'paragraph_id'

# --------------------------------------------------
# 0️⃣ Preparação: parse + explode
# --------------------------------------------------
cols_to_explode = [
    'human_action_part_name',
    'human_action_part_number',
    'human_action_chapter_name',
    'human_action_chapter_number'
]

df = mises_citations_df.copy()

def safe_literal_eval(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return [x]
    return x

# parse
for col in cols_to_explode:
    df[col] = df[col].apply(safe_literal_eval)

# garante listas
for col in cols_to_explode:
    df[col] = df[col].apply(
        lambda x: x if isinstance(x, list) else [x]
    )

# explode alinhado
df = df.explode(cols_to_explode)

# --------------------------------------------------
# 1️⃣ Filtragem (Mises)
# --------------------------------------------------
mises_filtered = df[
    (df['human_action_chapter_name'] != 'not found') &
    (df['human_action_part_number'] != 'WHOLE')
].copy()

# --------------------------------------------------
# 2️⃣ Tipagem (robusta)
# --------------------------------------------------
mises_filtered['human_action_part_number'] = pd.to_numeric(
    mises_filtered['human_action_part_number'],
    errors='coerce'
)

mises_filtered['human_action_chapter_number'] = pd.to_numeric(
    mises_filtered['human_action_chapter_number'],
    errors='coerce'
)

# remove NaNs
mises_filtered = mises_filtered.dropna(
    subset=['human_action_part_number', 'human_action_chapter_number']
)

mises_filtered['human_action_part_number'] = mises_filtered['human_action_part_number'].astype(int)
mises_filtered['human_action_chapter_number'] = mises_filtered['human_action_chapter_number'].astype(int)

# --------------------------------------------------
# 3️⃣ Co-citação via unidade (paragraph_id)
# --------------------------------------------------
hayek_units = set(hayek_citations_df[unit])

mises_filtered['co_cited_with_hayek'] = (
    mises_filtered[unit].isin(hayek_units)
)

# --------------------------------------------------
# 4️⃣ Agregação por PART
# --------------------------------------------------
part_stats = (
    mises_filtered
    .groupby(['human_action_part_name', 'human_action_part_number'])
    .agg(
        total_citations=(unit, 'count'),
        co_citations=('co_cited_with_hayek', 'sum')
    )
    .reset_index()
)

# --------------------------------------------------
# 5️⃣ Percentual
# --------------------------------------------------
part_stats['co_citation_pct'] = (
    part_stats['co_citations'] / part_stats['total_citations'] * 100
)

# --------------------------------------------------
# 6️⃣ Ordenação
# --------------------------------------------------
part_stats = part_stats.sort_values(
    by='human_action_part_number',
    ascending=False
)

# --------------------------------------------------
# 7️⃣ Cores
# --------------------------------------------------
part_colors = {
    1: 'lightblue',
    2: 'lightgreen',
    3: 'lightcoral',
    4: 'lightgoldenrodyellow',
    5: 'lightsalmon',
    6: 'lightpink',
    7: 'lightgray'
}

part_stats['bar_color'] = part_stats['human_action_part_number'].map(
    lambda x: part_colors.get(x, 'gray')
)

# --------------------------------------------------
# 8️⃣ Plot
# --------------------------------------------------
plt.figure(figsize=(12, 8))

plt.barh(
    part_stats['human_action_part_name'],
    part_stats['co_citation_pct'],
    color=part_stats['bar_color'],
    edgecolor='black'
)

plt.title('Percentual de co-citação com Hayek por parte de Human Action')
plt.xlabel('% de co-citação')
plt.ylabel('Part')

plt.tight_layout()
plt.savefig('../reports/human_action_co_citation_pct.png')
plt.show()

# --------------------------------------------------
# 9️⃣ Debug
# --------------------------------------------------
print("\nParts encontrados:")
print(part_stats['human_action_part_number'].unique())

print("\nResumo:")
print(part_stats[['human_action_part_name', 'co_citation_pct']])
```
