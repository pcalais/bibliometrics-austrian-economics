```python
import pandas as pd

refs_df = pd.read_csv("../data/processed/refs.csv")

refs_df.info()
```

```python
mises_refs_df = refs_df[refs_df.author == 'Mises'].sort_values(by='page')

mises_refs_df.head(100)
```

```python
import pandas as pd

# Lê o arquivo assumindo que é separado por vírgulas
human_action_page_ranges_df = pd.read_csv("chapter-ranges/human-action-page-to-chapter.txt")

human_action_page_ranges_df.head()
```

```python
def get_human_action_chapter_by_page(page):
    row = human_action_page_ranges_df[
        (human_action_page_ranges_df['start_page'] <= page) &
        (human_action_page_ranges_df['end_page'] >= page)
    ]
    if not row.empty:
        return int(row.iloc[0]['chapter'])
    else:
        return None  # ou você pode retornar "Desconhecido", -1, etc.

assert get_human_action_chapter_by_page(111)
```

```python
mises_refs_df['human_action_chapter_by_ref_page'] = mises_refs_df['page'].apply(get_human_action_chapter_by_page)
mises_refs_df['human_action_chapter_number'] = mises_refs_df['human_action_chapter_by_ref_page']


```

```python
chapter_to_part_df = pd.read_csv("chapter-ranges/chapter-to-part.csv")

chapter_to_part_df.head()
```

```python
import pandas as pd

# Exemplo de join usando merge
mises_refs_df = pd.merge(
    mises_refs_df,
    chapter_to_part_df,
    left_on='human_action_chapter_number',
    right_on='human_action_chapter_number',
    how='left'  # ou 'inner', 'right', 'outer', dependendo do seu objetivo
)
```

```python
mises_refs_df.to_csv("../data/processed/mises_refs.csv")
```
